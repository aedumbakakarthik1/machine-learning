```python
# importing the matplotlib library 

import matplotlib.pyplot as plt
```

```python
# importing the numpy library 

import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z= np.cos(x)
```

```python
print(y)
```

```python
print(z)
```

Ploting the data


```python
# sine wave
plt.plot(x,y)
plt.show()
```

```python
# cos wave
plt.plot(x,z)
plt.xlabel("x_axis - angle")
plt.ylabel("y_axis - sine value")
plt.title("sine wave")
plt.show()
```

```python
plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

```python
# parabola
x= np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,"r+")
plt.show()
```

```python
plt.plot(x,y,"b.")
plt.show()
```

```python
x=np.linspace(-5,10,100)
plt.plot(x,np.sin(x),"g-")
plt.plot(x,np.cos(x),"r--")
plt.show()
```

# bar plot

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
lan=["eng","Garman","Telugu","hindi"]
people =[100,50,150,145]
ax.bar(lan,people)
plt.xlabel("language")
plt.ylabel("no.of peoples")
plt.show()
```

# Pie Char

```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
lan=["eng","Garman","Telugu","hindi"]
people =[100,50,150,145]
ax.pie(people,labels=lan, autopct='%1.1f%%')
plt.show()
```

# scatter plot

```python
a=np.linspace(0,10,100)
x=np.sin(a)
y=np.cos(a)
fig2= plt.figure()
ax= fig2.add_axes([0,0,1,1])
ax.scatter(a,x,c="g")
ax.scatter(a,y,c="r")
```

# 3d scatter plot

```python
from matplotlib import projections
fig3= plt.figure()
ax= plt.axes(projection = '3d')
z=20 * np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap="Blues")
plt.show()
```
