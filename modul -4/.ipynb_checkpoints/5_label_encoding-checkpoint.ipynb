Label Encoding:
* converting the label into numeric form

```python
# import the dependencies 

import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

label Encoding of the breast cancer dataset

```python
df=pd.read_csv("/content/data.csv")
```

```python
df.head()
```

```python
# finding the count of different label
df.value_counts("diagnosis")
```

```python
# or
df["diagnosis"].value_counts()
```

```python
# load the label encoder function

label_encode =LabelEncoder()
```

```python
labels=label_encode.fit_transform(df.diagnosis)
```

```python
labels
```

```python
# appending to the dataFrame
df["target"]=labels
```

```python
df.head()
```

0 -> bengin (b)

1 ->malignant (m)


* note it will coder by alphabet order

```python
df["target"].value_counts()
```
